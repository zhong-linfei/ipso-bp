```python
import pandas as pd
import numpy as np

# 读取数据
data = pd.read_excel('./data/data1.xlsx')

# 标准化数据
data_norm = (data - data.mean()) / data.std()

# 计算相关系数矩阵
corr_matrix = np.corrcoef(data_norm, rowvar=False)

# 将相关系数矩阵转换为DataFrame格式
corr_df = pd.DataFrame(corr_matrix, columns=data.columns, index=data.columns)

# 打印相关系数矩阵
print(corr_df)
# 保存到文件中
corr_df.to_csv('./data/correlation.csv', index=False)
```
